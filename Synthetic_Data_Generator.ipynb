```python
import pandas as pd
df = pd.read_csv("/content/sample_data/MOCK_DATA.csv")
df.head()
```

```python
from sdv.metadata import Metadata
data_dict = {'data': df}
metadata = Metadata.detect_from_dataframes(data_dict)
```

```python
metadata.update_column(
    table_name='data',
    column_name="Transplant Center",
    sdtype = "secondary_address",
    pii=False)
metadata.update_column(
    table_name='data',
    column_name="COI Number",
    sdtype = "id",
    regex_format="^CX[0-9]{4}$")
metadata.update_column(
    table_name='data',
    column_name="Patient Name",
    sdtype = "name",
    pii=False)
```

```python
metadata
```

```python
from sdv.multi_table import HMASynthesizer

synthesizer = HMASynthesizer(metadata)
synthesizer.fit(data_dict)
synthetic_data = synthesizer.sample(scale=10000)
```
